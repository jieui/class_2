```python
# 결측치 중 성별 필드처리 : 최빈값으로 채우기(244~245p)
# df.drop(['성별'], axis=1, inplace=True) # axis=1 :열 삭제하려면 1로해야됨 / inplace=True:이거 적용하면 따로 변수에 할당하지 않아도 변경이 즉시 적용됨
# df.head()

df['식음메뉴명(영문)'].value_counts() # 최빈값 찾기
df['식음메뉴명(영문)'] = df['식음메뉴명(영문)'].fillna('Seasonal Fresh Fruits') #최빈값 Seasonal Fresh Fruits으로 대체
df.isnull().sum() #결측여부 확인

df['고객여권발급국가코드'].value_counts()
df['고객여권발급국가코드'] = df['고객여권발급국가코드'].fillna('CHN') #최빈값 CHN으로 대체
df.isnull().sum() #결측여부 확인

df['고객여권발급국가명'].value_counts()
df['고객여권발급국가명'] = df['고객여권발급국가명'].fillna('중국') #최빈값 중국으로 대체
df.isnull().sum() #결측여부 확인

# 주문수량 상위 10개보기 (242p)
df_sorted = df.sort_values('주문수량', ascending=False) #오름차순
# df_sorted
df_sorted.head(10)
```

```python
# value_counts 이용하여 데이터 셋 집계하기(243)
#  ->영업점명 별 이용 고객수
#  ->식음메뉴명(한글)별 이용 고객수
#  ->성별별 고객수
#  ->고객여권발급국가명 별 이용 고객수

a = df['영업점명'].value_counts()
print(a)
print()
b = df['식음메뉴명(한글)'].value_counts()
print(b)
print()
c = df['고객여권발급국가명'].value_counts()
print(c)
```

```python
# groupby를 이용하여 주문수량 집계하기(246)
#  ->영업점 별 주문수량 합, 평균 ->코엑스, 드래곤, 롯데 각각 주문수량과 평균 구하기 
#  ->식음메뉴명(한글) 별 주문수량 합
#  ->고객여권발급국가별 주문수량 합
#  ->영업점별 고객여권발급국가별 주문수량 합

import pandas as pd
df=pd.read_csv('C:\data\dataList\그랜드코리아레저(주)_영업점별 고객 식음료 주문 내역_20230630.csv')
# df.info()

sm=df['주문수량'].groupby(df['영업점명']).sum()
mn=df['주문수량'].groupby(df['영업점명']).mean() #df['주문수량'].groupby(df["영업점명"]).mean() 이렇게 하면 에러남 []여기 사이에 수치형데이터 들어가야됨 영업점명은 수치형 아니므로 안됨 
print(sm,'\n\n',mn)

sm_1=df['주문수량'].groupby(df['식음메뉴명(한글)']).sum()
mn_1=df['주문수량'].groupby(df['식음메뉴명(한글)']).mean()
print('\n', sm_1,'\n\n',mn_1)

sm_2=df['주문수량'].groupby(df['고객여권발급국가명']).sum()
mn_2=df['주문수량'].groupby(df['고객여권발급국가명']).mean()
print('\n', sm_2,'\n\n',mn_2)

df.groupby(['영업점명','고객여권발급국가명']).mean()
```
