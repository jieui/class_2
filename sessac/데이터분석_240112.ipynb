```python
#강사님의 보내주신 코드 

from glob import glob
import pandas as pd

fileList=glob('c:/data/dataList/happy/*.xlsx')
file=fileList[0]
df=pd.read_excel(file).iloc[:,1:4] #1,2,3열인 시도, 구군, 평균 읽어와 df에 저장
happy=pd.DataFrame({'시도': 
                          df['시도'].unique()}) # df['시도'].unique() : df의 '시도'열에서 고유한 값 추출(중복된 값 제외하고 유일한 값들 가져오는거)
                                                #{'시도': df['시도'].unique()} : 딕셔너리생성 / '시도' : Key,  df['시도'].unique(): Values
                                                #DataFrame <- 판다스에서 제공하는 라이브러리
#print(happy[:3])  #시도데이터를 그룹으로 제작



for file in fileList:
    print(file)
    colName=file.split('_')[1].split('.')[0] #파일명에서 건강, 안전 등 이름 분리
    df=pd.read_excel(file).iloc[:,1:4]
    df.columns=['시도','군구','평균'] # 삶의 만족도.xlsx 파일이 평균필드가 없어서 이름통일
    tmp=df['평균'].groupby(df['시도']).mean()
    happy=pd.merge(happy,tmp.rename(colName), on='시도')

happy.to_csv('c:/data/dataList/happy.csv',index=False, encoding='utf-8')

#메모리 제거하고 다시 자료 불러오는 작업을 권장함
import pandas as pd
df=pd.read_csv('c:/data/dataList/happy.csv')
df
```

```python
import numpy as np
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
```

```python
age=np.array([25,30,35,40,45, 50, 55, 60, 65, 70])
purchase_amount=np.array([100, 120, 80, 60, 90, 75, 110, 95, 130, 105])

import pandas as pd
data=pd.DataFrame({'나이':age, '금액':purchase_amount})
data.head()
```

```python
data=data.values
data
```

```python
# k=2
kmeans=KMeans(n_clusters=2)
kmeans.fit(data)

#각 데이터 포인트의 클러스터 할당
labels=kmeans.labels_

#각 클러스터의 중심 좌표
centers=kmeans.cluster_centers_
centers

#시각화
# plt.scatter(, , c=labels, cmap='viridis')
# plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], marker='X', s=200, c='red')
# plt.xlabel('나이')
# plt.ylabel('구매금액')
# plt.title('K-means Clustering')
# plt.legend()
# plt.show()
```
